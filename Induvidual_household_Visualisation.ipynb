```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "/content/resampled_15min_data.csv"
df = pd.read_csv(file_path)
```

```python
sns.set(style="whitegrid")
```

```python

for col in df.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], kde=True, bins=40)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
#df = df.drop(columns=['ID'])
plt.figure(figsize=(12, 8))
sns.heatmap(df.select_dtypes(include=['float64', 'int64']), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot of {col}')
    plt.show()
```

```python
sample_df = df.sample(min(500, len(df)))
sns.pairplot(sample_df, diag_kind='kde')
plt.show()
```

```python
for col in ['Energy_kWh_cummlative', 'Energy_kWh', 'apparent_power']:
    if col in df.columns:
        plt.figure(figsize=(8, 5))
        sns.violinplot(x=df['power_factor'], y=df[col], hue=df['power_factor'], palette='muted')
        plt.title(f'{col} Distribution by Mode')
        plt.show()
```
